```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import ast

class UniqueGraph():
    def __init__(self):
        self.input_values = ast.literal_eval(input('Enter the values: '))
        self.unique_values = []
        for a, b in self.input_values:
            self.unique_values.append(a and b)
        for i in list(range(1,10)):
            for a in self.unique_values:
                while self.unique_values.count(a) > 1:
                    self.unique_values.remove(a)
        self.unique_values.sort()
        self.result_dict = {}
        for first, second in self.input_values:
            self.result_dict.setdefault(first, []).append(second)
        self.reflexive_pairs = []
        for a, b in list(self.input_values):
            if a == b:
                self.reflexive_pairs.append((a,b))
        self.primes_list = []
        for num in self.unique_values:
            prime = True
            for i in range(2,num):
                if (num%i==0):
                    prime = False
            if prime:
               self.primes_list.append(num)
            
    def check_uniqueness(self):
        if len(self.reflexive_pairs) == len(self.unique_values):
            print(self.reflexive_pairs)
            print("Unique values check ✅")
            return True
        else:
            print("Unique values check ❌")
            return False
        
    def check_antisymmetry(self):
        antisymmetric_pairs = []
        for b in self.input_values:
            swap1 = b[0]
            swap2 = b[1]
            newtuple = (swap2, swap1)
            antisymmetric_pairs.append(newtuple)
        for ü in self.reflexive_pairs:
            if ü in antisymmetric_pairs:
                antisymmetric_pairs.remove(ü)
            else:
                None
        print(antisymmetric_pairs)
        for q in antisymmetric_pairs:
            if q in self.input_values:
                print("Anti-Symmetric check ❌")
                return False
        print("Anti-Symmetric check ✅")
        return True
    
    def check_transitivity(self):      
        for a, b in self.input_values:
            for x in self.result_dict[b]:
                if (x in self.result_dict[a]):
                    None
                else:
                    print("Transitive check ❌")
                    print("There is no ({},{}) in the {}".format(a, x, self.result_dict[a]))
                    return False
        print("Transitive check ✅")
        return True
    
    def draw_graph(self):
        pos = {}
        rand_list = list(range(1,len(list(self.result_dict.keys()))+1))
        for a in (list(self.result_dict.keys())):
            if a == 1:
                pos.setdefault(a, ((len(list(self.result_dict.keys()))/2), -len(list(self.result_dict.keys()))*2-4))
            elif a in self.primes_list:
                pos.setdefault(a, (a, -len(list(self.result_dict.keys()))*2))
            elif len(list(self.result_dict[a])) == 1:
                exit_r = random.choice(rand_list)
                pos.setdefault(a,(exit_r, 0))
                rand_list.remove(exit_r)
            else:
                exit_r = random.choice(rand_list)
                pos.setdefault(a, (exit_r, (-len(list(self.result_dict[a])))*2))
                rand_list.remove(exit_r)
        edges_dict = {}
        list1_reverse = list(self.unique_values)
        for a in list1_reverse:
            for b in list1_reverse:
                if (a%b==0) and (a != b):
                    edges_dict.setdefault(a, []).append(b)
        edge_list = [(x,y) for x,y in self.input_values if x!=y]
        for i in list(range(1,10)):
            for a, b in edge_list:
                if b in list(edges_dict.keys()):
                    for z in edges_dict[b]:
                        if (z!=a) and (z%a==0):
                            while (a,b) in edge_list:
                                edge_list.remove((a,b))
        T = nx.DiGraph()
        T.add_nodes_from(list(pos.keys()))
        T.add_edges_from(edge_list)
        plt.figure()
        if ( (self.check_uniqueness() == True) and (self.check_antisymmetry() == True) and (self.check_transitivity()== True) ):
            nx.draw(T, pos, node_color='black', node_size=600, font_size= 15, font_color='yellow', with_labels=True, arrowsize=18, edge_color='green')
        else:
            return "Conditions not satisfied."

unique_graph_instance = UniqueGraph()
result = unique_graph_instance.draw_graph()
if isinstance(result, str):
    print(result)
else:
    plt.show()  
```
